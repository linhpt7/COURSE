## Bài 2
Mở file `us_youtube_trending.xlsx`:
1. Xóa tất cả bản ghi có giá trị trống (NaN) trong cột `video_id`.
2. Hiển thị tất cả dữ liệu trùng lặp trong cột `video_id`.
3. Xóa trùng sao cho chỉ giữ lại 1 `video_id` có `trending_date` gần nhất.

```python
import numpy as  np
import pandas as pd
```

```python
yt = pd.read_excel(r'..\data\us_youtube_trending.xlsx')
yt.head()
```

#### Xóa giá trị thiếu

```python
yt = yt[~yt.video_id.isnull()]
yt
```

#### Hiển thị bản ghi trùng lặp

```python
yt[yt.duplicated('video_id', keep=False)].sort_values(by=['video_id'])
```

Giữ lại ngày gần nhất của mỗi `video_id`.

```python
yt.trending_date = pd.to_datetime(yt.trending_date)
```

```python
yt = yt.sort_values(by=['video_id', 'trending_date']).drop_duplicates('video_id', keep='last')
yt
```

```python
yt = yt.reset_index(drop=True)
yt
```

## Bài 3
Đọc dữ liệu `temperatures.xlsx` rồi tính nhiệt độ trung bình tháng, tuần của mỗi bang. Quy đổi nhiệt độ trong dataset từ độ F sang độ C bằng công thức: $x^{\circ}C = (y^{\circ}F-32) \times 5/9$

```python
import numpy as  np
import pandas as pd
```

```python
temp = pd.read_excel(r'..\data\temperatures.xlsx')
temp.head()
```

```python
# convert Fahrenheit to Celcius
temp.temperature = (temp.temperature - 32) * 5/9

# convert to datetime
temp.date = pd.to_datetime(temp.date)

# detach month and week
temp['month'] = temp.date.dt.to_period('M')
temp['week'] = temp.date.dt.strftime('%Y-%U')
```

```python
temp.head()
```

```python
temp.groupby(['city', 'month']).mean().reset_index()
```

```python
temp.groupby(['city', 'week']).mean().reset_index()
```
