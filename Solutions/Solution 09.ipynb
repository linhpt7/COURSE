## Bài 5
Unpivot dữ liệu dưới đây, sử dụng `append` thay vì `melt`.

```python
wide = pd.DataFrame({
    'Color': ['Red', 'Green', 'Blue'],
    '2020 Q1': [1000, 1500, 2000],
    '2020 Q2': [1200, 1500, 2200],
    '2020 Q3': [1500, 1575, 2000],
    '2020 Q4': [1700, 1800, 2800]
})
wide
```

```python
long = pd.DataFrame({
    'Color': [],
    'Value': [],
    'Quarter': []
})
long
```

```python
for i in wide.columns[1:]:
    to_append = wide[['Color', i]]
    to_append = to_append.rename(columns={i: 'Value'})
    to_append['Quarter'] = i
    
    long = long.append(to_append)
```
